# Construcción de Direcciones

```python
df['fiscal_codigo_postal'].dtypes
```

```python
# Pasar de decimal a entero
df['codigo_postal'] = pd.to_numeric(df['fiscal_codigo_postal'], errors='coerce')
df['codigo_postal']= df['codigo_postal'].fillna(0)
df['codigo_postal']= df['codigo_postal'].astype(int)
df['codigo_postal'].head()
```

```python
# Códigos postales errones
cp_erroneos = df[df['codigo_postal'] == 0]
print(cp_erroneos.shape)
cp_erroneos['fiscal_codigo_postal'].value_counts()
```

# Crear columna de dirección

```python
# df['concatenada'] = df[['col1', 'col2', 'col3']].astype(str).agg(lambda x: '-'.join(x.dropna()), axis=1)
df['direccion'] = df[['fiscal_estado_nombre', 
                      'fiscal_municipio_nombre', 
                      'fiscal_localidad_nombre', 
                      'codigo_postal',
                      'fiscal_asentamiento_nombre',
                      'fiscal_nombre_calle',
                      #'registro_razon_social',                     
                      ]].astype(str).agg(lambda x: ', '.join(x.dropna()), axis=1)
```
